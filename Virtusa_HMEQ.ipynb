```
#replacing with mean values
from statistics import mode

print("MEAN DATA\n")
mean_columns = ["MORTDUE","VALUE","DEBTINC","CLAGE"]

for i in mean_columns:
  #finding mean and replacing
  original_mean = data[i].mean()
  data[i] = data[i].fillna(original_mean)
  new_mean = data[i].mean()
  #data analysis after change
  print(i+" data analysis:")
  print(data[i].describe())
  print("Number of empty cells:",end='')
  print(data[i].isnull().sum())
  print("Deviation of ",i," is:",new_mean/original_mean,'\n')

#replacing with mode values

print("MEDIAN DATA\n")
median_columns = ["YOJ","CLNO","NINQ"]

for i in median_columns:
  #finding median and replacing
  original_median = data[i].median()
  data[i] = data[i].fillna(original_median)
  new_median = data[i].median()
  #data analysis after change
  print(i+" data analysis:")
  print(data[i].describe())
  print("Number of empty cells:",end='')
  print(data[i].isnull().sum())
  print("Deviation of ",i," is:",new_median/original_median,'\n')

#replacing with mode values

print("MODE DATA\n")
mode_columns = ["REASON","JOB","DEROG","DELINQ"]

for i in mode_columns:
  #finding mean and replacing
  original_mode = mode(data[i])
  data[i] = data[i].fillna(original_mode)
  new_mode = mode(data[i])
  #data analysis after change
  print(i+" data analysis:")
  print(data[i].describe())
  print("Number of empty cells:",end='')
  print(data[i].isnull().sum())
  print("Deviation of ",i," is:",new_mode,original_mode,'\n')
```

```
#3 - Jun -2020
#check it out

#data reading
# from google.colab import files
# uploaded = files.upload()

#data loading
import pandas as pd
import numpy as np

extVars_data = []

data = pd.read_csv('hmeq.csv')
# 1. Missing Data
#replacing with mean values
mean_columns = ["MORTDUE","VALUE","DEBTINC","CLAGE"]

mean_dict = dict()
for i in mean_columns:
  #finding mean and replacing
  original_mean = data[i].mean()
  data[i] = data[i].fillna(original_mean)
  mean_dict[i] = original_mean
extVars_data.append( mean_dict )

#replacing with median values
median_columns = ["YOJ","CLNO","NINQ"]

median_dict = dict()
for i in median_columns:
  #finding median and replacing
  original_median = data[i].median()
  data[i] = data[i].fillna(original_median)
  median_dict[i] = original_median
extVars_data.append( median_dict )

#replacing with mode values
from statistics import mode
mode_columns = ["REASON","JOB","DEROG","DELINQ"]

mode_dict = dict()
for i in mode_columns:
  #finding mean and replacing
  original_mode = mode(data[i])
  data[i] = data[i].fillna(original_mode)
  mode_dict[i] = original_mode
extVars_data.append( mode_dict )

# 1. Missing values End

# 2. Outliers
data_numeric = data.select_dtypes(include=[np.number]).columns.values
low = 0.005
high = 0.995
for col in data_numeric:
    if col not in ['BAD']:
      data[col] = np.where(data[col] < data[col].quantile(low), data[col].quantile(low),data[col])
      data[col] = np.where(data[col] > data[col].quantile(high), data[col].quantile(high),data[col])
# Outliers End

columns = [ 'BAD', 'LOAN', 'VALUE', 'MORTDUE', 'REASON', 'DEBTINC' ]
df = data[columns].copy()

# 3. Categorical Data
# Dummy for Reason Column
data['HOMEIMP'] = [1 if i == 'HomeImp' else 0 for i in data["REASON"]]
data.drop(['REASON'], axis=1, inplace=True)

# dictionary based on frequency (by default in sorted order)
job_frequency_dict = data['JOB'].value_counts().to_dict()
print(job_frequency_dict)
# Now replacing dictionary sorted frequencies with (0, 5)
counter = 0
for i in job_frequency_dict:
  job_frequency_dict[i] = counter
  counter += 1
# creating JOBIND with sorted values of JOB column
data['JOBIND'] = [job_frequency_dict[i] for i in data['JOB']]
data.drop(['JOB'], axis = 1, inplace = True)
# 3. Categorical End

# 4. Feature Scaling
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# scaled_data = scaler.fit_transform(data.iloc[:,1:])
# 4. End Feature Scaling

print(data.columns)
# print(scaled_data)

# 5. Dependent and Independent Split 
y = data.iloc[:,0]
x = data.iloc[:,1:]
# 5. Dependent and Independent Split End

# 6. Train Test Split
from sklearn.model_selection import train_test_split 
xtrain, xtest, ytrain, ytest = train_test_split( x, y, test_size = 0.25, random_state = 0) 
# 6. Train Test Split End

# 7. Model Training
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
model = RandomForestClassifier(criterion = 'gini', max_depth = 75, n_estimators = 125, n_jobs=-1, warm_start=True)
model.fit(xtrain, ytrain.ravel())
extVars_data.append( model )
# 7. Model Training End

# 8. Saving trained Model
import pickle
with open( "extVars.pkl", "wb" ) as f:
    pickle.dump( extVars_data, f )
# 8. End Saving Trained Model

y_pred = model.predict(xtest)

#accuracy check
accu_score = accuracy_score(ytest,y_pred)
print('Accuracy: ', (accu_score*100).round(5),'%')
import eli5

# eli5.show_prediction(model, xtest.iloc[0], feature_names=xtest.columns.tolist(), show_feature_values=True)
# eli5.explain_prediction(model, xtest.iloc[0], target_names= ytrain.unique())
explaination_pred = eli5.explain_prediction_df(estimator=model, doc=xtest.iloc[0]).round( { 'value' : 3 } )
explaination_pred = explaination_pred[explaination_pred.feature != '<BIAS>'].reset_index( drop=True )

print( explaination_pred )

pred_dict = dict()
pred_dict['target'] = 0
pred_dict['features'] = [ i for i in explaination_pred['feature'].tolist() ]
pred_dict['values'] = explaination_pred['value'].tolist()

print( pred_dict['features'] )

explaination_pred = explaination_pred.iloc[:6]

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure( figsize = (10,5) )

ax = sns.barplot(x='weight', y='feature', data=explaination_pred)

plt.title("Top Features", fontsize = 30, loc = 'center', pad=10,
            fontdict = { 'family': 'serif', 'color': 'darkred', 'weight': 'normal', 'size': 25 })
plt.xlabel('Importance',
        fontdict = { 'family': 'serif', 'color': 'darkred', 'weight': 'normal', 'size': 16 })
plt.ylabel('Features',
        fontdict = { 'family': 'serif', 'color': 'darkred', 'weight': 'normal', 'size': 16 })
plt.show()
# import eli5\neli5.formatters.as_dataframe.explain_prediction_df(model, xtest.iloc[0],target_names= ytrain.unique())\n# eli5.formatters.format_as_dict(model)\n# eli5.formatters.image.format_as_image.explain_prediction_df(model, resampling_filter=Image.LANCZOS, colormap=matplotlib.cm.viridis, alpha_limit=0.65)\n# eli5.expand_heatmap(heatmap, image)\neli5.formatters.image.format_as_image(model, xtest.iloc[0],target_names= ytrain.unique())
```
